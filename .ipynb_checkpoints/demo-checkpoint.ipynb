```python
import transformers
import numpy as np
import torch
from transformers import AutoModelWithLMHead, AutoTokenizer
from visual_t5 import VisualT5

from PIL import Image
import matplotlib.pyplot as plt
from torch.nn.functional import softmax
```

```python
vqa_model_path = '.././results/visual_qa_st_b128_p20/checkpoint-35000/'

tokenizer = AutoTokenizer.from_pretrained('t5-small')

vqa_model = VisualT5.from_pretrained(vqa_model_path)
vqa_model = vqa_model.cuda()
```

```python
def make_img_input(input_):
    max_input_len = 64
    num_regions = 36
    input_text = input_
    
    inputs = tokenizer.encode(input_text, add_special_tokens=False, return_tensors="pt")
        
    len_tokens= inputs.size(1)
    img_prefix_token = torch.tensor([1023,10]).repeat(inputs.size(0), 1) # 'image :' in T5
    pads = torch.zeros((1, max_input_len-inputs.size(1)),dtype=torch.long)  
    inputs = torch.cat((inputs, pads), axis=1)
    inputs = torch.cat((inputs, img_prefix_token), axis=1)

    attention_mask = torch.ones(1, (inputs.size(1) + num_regions), dtype=torch.long)                   
    attention_mask[0][len_tokens:max_input_len] = 0
    
    return inputs, attention_mask

def generate_wt_img_batch(vqa_model, inputs_, img_feature_, beam_size=1):
    if(len(inputs_)==0):
        return [], []
    
    for i in range(len(inputs_)):
        if(i == 0):
            inputs, attention_mask = make_img_input(inputs_[i]) 
        else:
            inputs_e, attention_mask_e = make_img_input(inputs_[i])
            inputs = torch.cat((inputs, inputs_e), axis=0)
            attention_mask = torch.cat((attention_mask, attention_mask_e), axis=0)

    inputs = inputs.cuda()
    attention_mask = attention_mask.cuda()
    img_feature_ = img_feature_.repeat(len(inputs_), 1, 1).cuda()
    
    model_kwargs = {}
    with torch.no_grad():
        encoder_output = vqa_model.encoder(input_ids=inputs, 
                           attention_mask=attention_mask, 
                           img_features = img_feature_,
                                            return_dict=True)
        model_kwargs["attention_mask"] = attention_mask    
        model_kwargs["encoder_outputs"] = encoder_output


        outputs_ = vqa_model.generate(do_sample=False, output_scores=True,
                                           return_dict_in_generate=True, num_beams=1, **model_kwargs)
     
    outputs = outputs_['sequences']
    unans_prob = (softmax(outputs_['scores'][0], dim=1)[:,73].cpu().numpy()) # 73 is 'uns_'
    ans_prob = (1 - unans_prob)
    ans =  [tokenizer.decode(x).replace('<pad>', '').replace('</s>', '')[1:] for x in outputs][0]
    return ans, ans_prob
```

```python
img_f = 'sample.jpg'
im = Image.open(img_f)
plt.imshow(im)

question = 'What kind of animals are the men on?'

img_npy = torch.tensor(np.load('sample.npy')).type('torch.FloatTensor')
ans, ans_prob = generate_wt_img_batch(vqa_model, [question], img_npy)

print('Ques: %s'%question)
print('Pred: %s'%ans)
```
